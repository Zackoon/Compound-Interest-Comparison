```python
#Import statements pasted from Data 8 Homework

import numpy as np
from datascience import *
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import warnings
warnings.simplefilter('ignore', FutureWarning)
```

```python
def one_year_interest(amount, yearly_interest):
    '''Returns one year of interest applied to the amount inputted'''
    return amount*yearly_interest
```

```python
#set up: 1 coin maximum in the "C" account at a rate of 6% annually.
#Each additional coin in C provides a 3.51% annual return. 
#Nexo on the other hand will provide a 5% annual rate no matter what.
#C is considered "safer" than "N", but we'll be checking whether the difference in returns is large enough
#to justify putting it in N anyways. 
#They will have 1 coin in C for the 6% returns, and is deciding which account to put the remaining 1.26
#In this case, only interest from the previous years will be changing the balance 
c_interest =  1.0351
n_interest = 1.05
c_amount = make_array(1.26)
n_amount = make_array(1.26)

for i in np.arange(15):
    c_amount = np.append(c_amount, 
                                one_year_interest(c_amount.item(i), c_interest+0.06))
    #0.06 comes from the interest in the C account for 1 coin. This is transferred after interest is calculated for the year.
    n_amount = np.append(n_amount, 
                                one_year_interest(n_amount.item(i), n_interest+0.06))
coin_table = Table().with_columns("Year", np.arange(0,16), "C", c_amount, "N", n_amount,
                                    "Difference", n_amount - c_amount)
coin_table.show()
coin_table.plot("Year") 
```
